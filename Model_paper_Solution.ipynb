```python
dataset['Engine']=dataset.transform(lambda x: x.fillna(x.median()))['Engine']
dataset['Power']=dataset.transform(lambda x: x.fillna(x.median()))['Power']
dataset['Seats']=dataset.transform(lambda x: x.fillna(x.median()))['Seats']
dataset['Mileage']=dataset.transform(lambda x: x.fillna(x.median()))['Mileage']
dataset['New_Price']=dataset.transform(lambda x: x.fillna(x.median()))['New_Price']
```

```python
# 12) Consider the Customer_Purchases.csv dataset and perform the following operations.
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
# Importing the dataset 
dataset = pd.read_csv('Customer_Purchases.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```python
dataset.head()
```

```python
dataset.shape
```

```python
# Spliting the dataset into the training dataset
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
# Feature Scaling 
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
# Training the Decision Tree Classification model on the Training set
# a) Criterion as Entropy
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(criterion = 'entropy', random_state = 0, max_depth = 3)
classifier.fit(X_train, y_train)
```

```python
# Predicting a new result
print(classifier.predict(sc.transform([[19,19000]])))
```

```python
# Predicting the test set result
y_pred = classifier.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
# Making Confusion matrix (Accuracy)
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```

```python
# b) Criterion as gini
classifier = DecisionTreeClassifier(criterion = 'gini', random_state = 0,  max_depth = 3)
classifier.fit(X_train, y_train)
```

```python
# Predicting a new result
print(classifier.predict(sc.transform([[35,20000]])))
```

```python
# Predicting the test set result
y_pred = classifier.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
# Making Confusion matrix (Accuracy)
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```
